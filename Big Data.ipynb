```python
superstore['Sales_r'] = round(superstore['Sales'], 2)
superstore['Profit_r'] = round(superstore['Profit'], 2)
cols = ['Sales','Profit']
superstore.drop(cols, axis=1, inplace=True)
```

```python
insignificant=['Row ID','Order ID','Ship Mode','Customer Name','Customer_no','Country','State_no','Postal Code','Region_no']
superstore.drop(insignificant, axis=1, inplace=True)
```

```python
returned_orders = superstore[superstore['Returned'] == True]
superstore = superstore.drop(returned_orders.index)
```

```python
cols=['Quantity','Sales_r','Profit_r']
plt.figure(figsize=[10,5])
sns.set(style="whitegrid")
ax = sns.boxplot(data=superstore[cols])
plt.title('Box Plot To Find Outliers')
plt.show()
```

```python
numeric_cols = superstore[['Sales_r', 'Profit_r', 'Quantity']]
Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1
Outliers = np.any((numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR)), axis=1)
print("Number of outliers detected:", Outliers.sum())
superstore = superstore[~Outliers]
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
x = superstore[['Sales_r', 'Quantity', 'Discount']]
y = superstore['Profit_r']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print ("Coefficient: ", model.coef_)
print("Intercept: ", model.intercept_)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print('r^2 score:', r2)
print('Mean Squared Error:', mse)
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Actual Profit')
plt.ylabel('Predicted Profit')
plt.title('Linear Regression Model: Actual vs Predicted Profit\nR-squared = {:.2f}'.format(r2))
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np
import pandas as pd
```

```python
x = superstore[['Sales_r', 'State', 'Sub-Category', 'Segment', 'Quantity']]
y = superstore['Profit_r']
x = pd.DataFrame(x)
preprocessor = ColumnTransformer(transformers=[('State', OneHotEncoder(),['State']),
                                               ('SubCat', OneHotEncoder(), ['Sub-Category']),
                                               ('Seg', OneHotEncoder(),['Segment'])], remainder='passthrough')
x = preprocessor.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print("Coefficient: ", model.coef_)
print("Intercept: ", model.intercept_)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print('R^2 Score:', r2)
print('Mean Squared  Error:', mse)
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Actual Profit')
plt.ylabel('Predicted Profit')
plt.title('Linear Regression Model: Actual vs. Predicted Profit\nr-squared = {:.2f}'.format(r2))
plt.show()
```
